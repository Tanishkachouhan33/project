```python
file2=pd.read_csv('/content/clean_data_second.csv')
```

```python
file2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X=file2[["views","likes","dislikes","comment_count","trend_day_count"]]
Y=file2['tag_appeared_in_title']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error # Using mean_squared_error for regression evaluation

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)
model = LogisticRegression() # Changed to LogisticRegression
model.fit(X_train, Y_train)
```

```python
predicted=model.predict(X_test)
predicted
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(Y_test,predicted)
print(a)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
cm = confusion_matrix(Y_test,predicted)


plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
report = classification_report(Y_test,predicted, output_dict=False)
print(report)
```
